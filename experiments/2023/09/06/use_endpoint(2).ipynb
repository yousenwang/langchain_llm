# Embedding

# LLM

```python
out2 = json.loads(r2.content.decode('utf-8'))
```

```python
out2["relavant_docs"]
```

```python

content2 = " ".join([cur["page_content"] for cur in out2["relavant_docs"]])
```

```python
content2
```

```python
%%time
r2_llm = requests.post(
    url='http://127.0.0.1:8000/completion',
    json = {
        "context": content2,
        "question": out2["user_input"]
    }
    )
```

```python
r2_llm
```

```python
r2_llm_out = json.loads(r2_llm.content.decode('utf-8'))
```

```python
r2_llm_out
```

```python
%%time
r2_llm = requests.post(
    url='http://127.0.0.1:8000/completion',
    json = {
        "context": content2,
        "question": out2["user_input"],
        "temperature": 1
    }
    )
```

```python
r2_llm_out = json.loads(r2_llm.content.decode('utf-8'))
```
